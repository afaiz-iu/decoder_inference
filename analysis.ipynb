```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('inference_times_seq.csv')

seq_len = data['seq_len']
mean_inf_time = data['mean_inf_time']
std_dev_inf_time = data['std_dev_inf_time']
```

```python
# bar graph with error bars
plt.figure(figsize=(12, 8))
plt.bar(seq_len, mean_inf_time, yerr=std_dev_inf_time, alpha=0.7, color='b', label='Mean Inference Time')
plt.xlabel('Sequence Length')
plt.ylabel('Mean Inference Time (ms)')
plt.title('Mean Inference Time vs Sequence Length')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xticks(seq_len)  
```

```python
degree = 2
coefficients = np.polyfit(seq_len, mean_inf_time, degree)
polynomial = np.poly1d(coefficients)

# x-values
x_poly = np.linspace(min(seq_len), max(seq_len), 100)
y_poly = polynomial(x_poly)

plt.plot(x_poly, y_poly, 'r-', label=f'{degree} Degree Polynomial Fit')
plt.legend()

plt.show()
```

```python
bar_width = 0.4  # Adjust this value as needed

# Plotting the bar graph with error bars
plt.figure(figsize=(14, 8))
plt.bar(seq_len, mean_inf_time, yerr=std_dev_inf_time, alpha=0.7, color='b', label='Mean Inference Time', width=bar_width)
plt.xlabel('Sequence Length')
plt.ylabel('Mean Inference Time (ms)')
plt.title('Mean Inference Time vs Sequence Length')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xticks(seq_len) 

degree = 1
coefficients = np.polyfit(seq_len, mean_inf_time, degree)
polynomial = np.poly1d(coefficients)

x_poly = np.linspace(min(seq_len), max(seq_len), 100)
y_poly = polynomial(x_poly)

plt.plot(x_poly, y_poly, 'r-', label=f'{degree} Degree Polynomial Fit')
plt.legend()

plt.show()
```
